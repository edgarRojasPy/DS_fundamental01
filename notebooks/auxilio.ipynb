```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ejemplo de datos
data = {'edad': [22, 25, 29, 25, 27, 30, 29, 30, 31, 29, 28, 22, 23, 25]}
df = pd.DataFrame(data)

# Histograma
plt.figure(figsize=(8, 5))
sns.histplot(df['edad'], bins=8, kde=True)  # kde=True agrega una línea de densidad
plt.title('Histograma de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```
